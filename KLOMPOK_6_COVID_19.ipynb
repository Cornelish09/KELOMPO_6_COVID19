```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# **Load Data**

```python
path = "covid_19_indonesia_time_series_all.csv"
df = pd.read_csv(path)
print(df.shape)
df.head()
```

# **Basic information about the dataset**

```python
df.info()
df.describe(include='all').T
```

# **Cek nilai duplikat & nilai unik**

```python
print("Jumlah baris duplikat:", df.duplicated().sum())
df.nunique(dropna=False).sort_values(ascending=False)
```

# **Visualisasikan jumlah nilai unik**

```python
uniq = df.nunique(dropna=False).sort_values(ascending=False).head(12)
uniq.plot(kind='bar', figsize=(8,4), title='Unique Count per Column (Top 12)')
plt.xlabel('Kolom'); plt.ylabel('Unique Count'); plt.tight_layout(); plt.show()
```

# **Menemukan null values**

```python
null_count = df.isna().sum().sort_values(ascending=False)
null_pct   = (df.isna().mean()*100).sort_values(ascending=False)
pd.concat([null_count.rename('null_count'), null_pct.rename('null_%')], axis=1)
```

# **Replace semua null values**

```python
df_filled = df.copy()
for c in df_filled.columns:
    if pd.api.types.is_numeric_dtype(df_filled[c]):
        df_filled[c] = df_filled[c].fillna(df_filled[c].median())
    else:
        mode = df_filled[c].mode(dropna=True)
        df_filled[c] = df_filled[c].fillna(mode.iloc[0] if not mode.empty else "Unknown")

df_filled.isna().sum().head(10)
```

# **Mengetahui tipe data**

```python
df_filled.dtypes.to_frame('dtype')
```

# **Filter data (contoh: level provinsi tahun 2021, DKI Jakarta)**

```python
dff = df_filled.copy()

cols = {c.lower(): c for c in dff.columns}
col_date   = cols.get('date') or cols.get('tanggal') or 'Date'
col_level  = cols.get('location_level') or 'Location_Level'
col_loc    = cols.get('location') or cols.get('province') or 'Location'

dff[col_date] = pd.to_datetime(dff[col_date], errors='coerce')

mask = pd.Series(True, index=dff.index)
if col_level in dff: mask &= dff[col_level].str.lower().eq('province')
mask &= dff[col_date].dt.year.eq(2021)
if col_loc in dff:   mask &= dff[col_loc].str.lower().isin(['dki jakarta','jakarta'])

dff = dff[mask].copy()
print(dff.shape)
dff.head()
```

# **Box plot (subset kolom numerik)**

```python
num_cols = [c for c in dff.columns if pd.api.types.is_numeric_dtype(dff[c])]
cand = ['New Cases','New Deaths','New Recovered','Total Cases','Total Deaths','Total Recovered']
take = [c for c in cand if c in dff.columns]
if not take:
    take = num_cols[:5]

if take:
    dff[take].plot(kind='box', figsize=(8,4), title='Boxplot (subset numerik)')
    plt.tight_layout(); plt.show()
else:
    print("Tidak ada kolom numerik yang cocok untuk boxplot.")
```

# **Correlation**

```python
corr = df.select_dtypes(include=np.number).corr(method='pearson')
corr.round(2)

# Heatmap sederhana (tanpa seaborn)
plt.figure(figsize=(8,6))
plt.imshow(corr, aspect='auto')
plt.colorbar(label='Pearson r')
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.index)), corr.index)
plt.title('Correlation Heatmap (numeric)')
plt.tight_layout(); plt.show()
```
